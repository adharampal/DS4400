Problem 4: 
Generate at random 3 matrices of size $3 \times 3$ and fill each entry with a random integer chosen from -10 to 10. Use an existing package to compute the inverse of each matrix if it exists. Include in your report:
- Each of the 3 matrices
- The inverses computed with the package
- Compute the product of each matrix with its inverse to check that you obtain the identity matrix.

```python
import numpy as np
import pandas as pd
```

```python
# 3 Random Matrices
A = np.random.randint(-10, 11, size=(3,3))
B = np.random.randint(-10, 11, size=(3,3))
C = np.random.randint(-10, 11, size=(3,3))

print('Matrix A')
print(A)

print('\nMatrix B')
print(B)

print('\nMatrix C')
print(C)

```

```python
# Computing Inverse

# To avoid scientific notation, I am rounding to 4 decimal points
print('Inverse of A')
A_inv = np.linalg.inv(A)
print(np.round(A_inv, 4))

print('\nInverse of B')
B_inv = np.linalg.inv(B)
print(np.round(B_inv, 4))

print('\nInverse of C')
C_inv = np.linalg.inv(C)
print(np.round(C_inv, 4))
```

```python
# Matrix Products
A_prod = np.round(A @ A_inv, 4)
print('Product of A and Inv:')
print(A_prod)

B_prod = np.round(B @ B_inv, 4)
print('\nProduct of B and Inv:')
print(B_prod)

C_prod = np.round(C @ C_inv, 4)
print('\nProduct of C and Inv:')
print(C_prod)
```

```python
# Cleaning the Data 
unclean_df = pd.read_csv('kc_house_data.csv')
data = unclean_df.drop(columns=['id', 'date', 'zipcode'])

# Getting the stats
stats = data.agg(['mean', 'min', 'max', 'var'])
print(stats)

lowest_mean_feature = stats.loc['mean'].idxmin()
lowest_mean = stats.loc['mean'].min()

print("\nLowest mean:", lowest_mean_feature)
print("Value:", lowest_mean)

highest_mean_feature = stats.loc['mean'].idxmax()
highest_mean = stats.loc['mean'].max()

print("\nHighest mean:", highest_mean_feature)
print("Value:", highest_mean)

lowest_var_feature = stats.loc['var'].idxmin()
lowest_var = stats.loc['var'].min()

print("\nLowest variance:", lowest_var_feature)
print("Value:", lowest_var)

highest_var_feature = stats.loc['var'].idxmax()
highest_var = stats.loc['var'].max()

print("Highest variance:", highest_var_feature)
print("Value:", highest_var)
```
